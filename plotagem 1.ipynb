```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs  
import cartopy.feature as cfeature
import numpy as np
```

```python
fig = plt.figure(figsize=(10, 8)) #criar uma figura no geral e dps plotar as coisas nela 
ax = fig.add_subplot(111, projection=ccrs.Robinson()) #plotar de fato a projeção escolhida
ax.add_feature(cfeature.LAND)# O comando feature é para adicionar coisas no mapa, como as delimitações etc
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.LAKES, edgecolor='red')
ax.add_feature(cfeature.RIVERS)

ax.gridlines(crs=ccrs.Robinson())# Adicionar linhas de grade
ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=2, color='gray', alpha=0.5, linestyle='--')#colocar lat e lom
ax.set_title('Projeção de Robinson', fontsize=15, y=1.07) # colocar um titulo e dar espaçamento do titulo pro mapa
```

```python
import matplotlib.ticker as mticker
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER
fig = plt.figure(figsize=(8, 6)) #criar uma figura DEFAULT e dps plotar em cima dela
ax = fig.add_subplot(111, projection=ccrs.PlateCarree()) #plotar de fato a projeção escolhida

ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)

ax.set_extent([-48, -34, 0, -18], ccrs.PlateCarree()) #Definir a extensão do mapa

states = cfeature.NaturalEarthFeature(category='cultural', name='admin_1_states_provinces_lines', scale='50m', facecolor='none')
ax.add_feature(states, edgecolor='black')


#insert caption in a box
ax.text(0.5, -0.15, 'Longitude', transform=ax.transAxes, ha='center', fontsize=12)
#insert caption in a box
ax.text(-0.20, 0.5, 'Latitude', transform=ax.transAxes, va='center', rotation='vertical', fontsize=12)

#inserindo linhas de grade 

g1 = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, linestyle='--')

#Apagando latitude e longitude das laterais caso necessario
g1.right_labels = False
g1.top_labels = False

#rotacionar o texto das legendas
g1.xlabel_style = {'rotation': 45}
g1.ylabel_style = {'rotation': 45}

#plotar a imagem de fundo
ax.stock_img() 
```

```python
import matplotlib.ticker as mticker
import numpy as np
from cartopy.mpl.gridliner import LATITUDE_FORMATTER, LONGITUDE_FORMATTER

fig = plt.figure(figsize=(8, 6)) #criar uma figura no geral e dps plotar as coisas nela 
ax = fig.add_subplot(111, projection=ccrs.PlateCarree()) #plotar de fato a projeção escolhida

ax.add_feature(cfeature.COASTLINE)# Adicionando as delimitações das costas
ax.add_feature(cfeature.BORDERS)# Adicionando Bordas

ax.set_extent([-90, -30, 10, -40], ccrs.PlateCarree()) #definir a extensão do mapa

states = cfeature.NaturalEarthFeature(category='cultural', name='admin_1_states_provinces_lines', scale='50m', facecolor='none')#tamanho cor e etc...
ax.add_feature(states, edgecolor='black')

#inserindo os dados 
ax.scatter(df['LONGITUDE'], df['LATITUDE'], s=50, color='red', transform=ccrs.PlateCarree())

#inserindo linhas de grade 

g1 = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, linestyle='--')

#Apagando latitude e longitude das laterais caso necessario
g1.right_labels = False
g1.top_labels = False

g1.ylocator = mticker.FixedLocator(np.arange(-40, 20, 10))

g1.yformatter = LATITUDE_FORMATTER
g1.xformatter = LONGITUDE_FORMATTER

g1.xlabel_style = {'size': 15}
g1.ylabel_style = {'size': 15}
#plotar a imagem de fundo
ax.stock_img() 
```
